## For loop: 

```python
import pandas as pd
import requests
import json
import datetime
import time
import numpy as np 
```

```python
# BDay is business day, not birthday...
from pandas.tseries.offsets import BDay

today = datetime.datetime.today()
ten_biz_days_before = today - BDay(10)

from_date = ten_biz_days_before.strftime("%d-%m-%Y")
to_date = today.strftime("%d-%m-%Y")

print(from_date)
print(to_date)
```

```python
df = pd.read_csv("stocklist.csv")
stock_list = df["SYMBOL"].tolist()
df.head(5)
```

```python
stock_list
```

```python
def test_abnormal(list_data):
    q75, q25 = np.percentile(list_data[:-1], [75 ,25])
    iqr = q75 - q25
    threshold = q75 + 1.75 * iqr
    last_value = list_data[-1]
    abnormal = False
    if last_value > threshold:
        abnormal = True
    else: 
        abnormal = False

    return abnormal
```

```python
# List of URLs
url1 = "https://www.nseindia.com/report-detail/eq_security"

# Header
header1 = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36"
}


s = requests.Session()
s.headers.update(header1)
r1 = s.get(url1)
s.headers.update(header1)


result = []
df3 = pd.DataFrame([])

for stock in stock_list: 
    print("Processing stock: ",stock)
    stock_name = stock
    url2 = "https://www.nseindia.com/api/historical/securityArchives?from="+from_date+"&to="+to_date+"&symbol="+stock_name+"&dataType=deliverable&series=EQ"
    r2 = s.get(url2, cookies=r1.cookies)
    time.sleep(1)
    if r2.status_code == 200:
        delivery_data = json.loads(r2.text)
        delivery_data = delivery_data.get('data') 
        if len(delivery_data)==0:
            result.append({'Stock Symbol': stock_name, 'Delivery to Trade Percentage' : '-1', 'Date' : 'Err', 'High Delivery' : 'Err' }) 
        else: 
            #print(delivery_data)
            df2 = pd.DataFrame.from_dict(delivery_data, orient='columns')
            df2.drop(['_id', 'COP_SR_NOS', 'COP_TRADED_QTY','TIMESTAMP','COP_DELIV_QTY','createdAt','updatedAt', 'mTIMESTAMP', '__v'], axis=1, inplace=True)
            report_date = df2["COP_TRADED_DT"].tolist()[-1]  # Last element in list 
            delivery_pct = df2["COP_DELIV_PERC"].tolist()[-1]  # Last element in list

            delivery_pct_list = df2["COP_DELIV_PERC"].tolist() 

            if test_abnormal(delivery_pct_list) : 
                result.append({'Stock Symbol': stock_name, 'Delivery to Trade Percentage' : delivery_pct, 'Date' : report_date, 'High Delivery' : 'Yes' })
            else: 
                result.append({'Stock Symbol': stock_name, 'Delivery to Trade Percentage' : delivery_pct, 'Date' : report_date, 'High Delivery' : 'No' })
  
    else:
        result.append({'Stock Symbol': stock_name, 'Delivery to Trade Percentage' : '-1', 'Date' : 'Err', 'High Delivery' : 'Err' })

    
df3 = pd.DataFrame.from_dict(result, orient='columns')


```
